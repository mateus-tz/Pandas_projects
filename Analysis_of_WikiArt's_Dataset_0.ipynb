```python
from matplotlib import ticker
x = df_date_number_of_picures['date'].values
y  = df_date_number_of_picures['number'].values

f, ax = plt.subplots(figsize = (25,20))

#ax.stem(df_date_number_of_picures['date'].values,df_date_number_of_picures['number'].values, linefmt = ':', use_line_collection = True )

markerline, stemlines, baseline = ax.stem(x , y, linefmt ='k-', basefmt = ' ' , bottom = 1.1, use_line_collection = True) 
#markerline.set_markerfacecolor('none') 

#markerline, stemlines, baseline = ax.stem(x, y, markerfmt='o', label='sin')
plt.setp(stemlines, 'color', plt.getp(markerline,'color'), alpha = .3, color = '#8f8f8f', linewidth = 0.6)
plt.setp(stemlines, 'linestyle', '-')
plt.setp(markerline, markersize = 10)
plt.setp(markerline, 'color', 'k')
markerline.set_markerfacecolor('#d1d1d1')

xticks = np.arange(1000, 2020, 100)
yticks = np.arange(0, 1300, 100)

ax.set_xticks(xticks)
ax.set_xticklabels(xticks, fontsize = 25, rotation = 0)

ax.set_yticks(yticks)
ax.set_yticklabels(yticks, fontsize = 25)

ax.set_ylim(-5, 1250)
ax.set_xlim(1015, 2030)

ax.tick_params(which = 'both', length = 10, width = 1.1)

ax.spines[['right', 'top']].set_visible(False)
ax.yaxis.set


#ax.set_title('')
ax.set_ylabel('Numero de obras artísticas', fontsize =  42)
ax.set_xlabel('Tempo(ano)',labelpad = 18 , fontsize = 42)

ax.xaxis.set_minor_locator(ticker.MultipleLocator(50))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(50))

#f.tight_layout()

#ax.set_yscale("log")
#f.savefig('/home/mateus/projects/ordinal_distribution_art/plot_number_time_of_artworks_2.pdf',  bbox_inches = 'tight')
```
